```python
#Reading CSV
import pandas as pd
import numpy as np
df=pd.read_csv("world_literacy_data_2021_2023.csv")
print(df)
```

```python
#Printing size of dataset(rows,columns)
print(df.shape)
print(df.size)
```

```python
#Cheking if there is a null values present or not
print(df.isnull().sum())
```

```python
#Check for the NULL values in dataset and if found replace the NULL values with the average values
df['GDP']=df['GDP'].replace({'\$':'',',':''},regex=True)
df['GDP']=pd.to_numeric(df['GDP'])
print(df)
```

```python
df['Literacy Rate']=df['Literacy Rate'].replace({'%':''},regex=True)
df['Literacy Rate']=pd.to_numeric(df['Literacy Rate'])
df
```

```python
#replace the NULL values with the average values
df.fillna(df.mean(numeric_only=True), inplace=True)

#print the updated dataset
df
```

```python
#Display the top 10 countries having highest GDP year 2023
top_gdp_2023 = df[df['Year'] == 2023].sort_values(by='GDP', ascending=False).head(10)
print(top_gdp_2023[['Country', 'GDP' , 'Year']])
```

```python
#Display top 10 countries having highest Literacy rate year 2023
top_literacy_2023 = df[df['Year'] == 2023].sort_values(by='Literacy Rate', ascending=False).head(10)
print(top_literacy_2023[['Country', 'Literacy Rate' , 'Year']])
```

```python
#Plot the year wise literacy rate of each country
import pandas as pd
import plotly.express as px

df =  pd.read_csv("world_literacy_data_2021_2023.csv")

# Create the Plotly figure
fig = px.line(df, x='Year', y='Literacy Rate', color='Country')

# Customize the layout
fig.update_layout(
    title='Year-wise Literacy Rate of Each Country',
    xaxis_title='Year',
    yaxis_title='Literacy Rate',
    legend_title='Country'
)

# Show the figure
fig.show()
```
